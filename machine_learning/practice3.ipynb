```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("/Users/navkiratsingh/Documents/coding/carprices.csv")
df.rename(columns={"Car Model":"carmodel"} ,inplace=True)
df
```

```python
from pandas import get_dummies
dumm=pd.get_dummies(df.carmodel)
dumm
dumm.replace(False,0, inplace=True)
dumm.replace(True,1, inplace=True)
dumm
```

```python
merged=pd.concat([df,dumm],axis='columns')
finaldf=merged.drop(['carmodel','Audi A5'],axis='columns')
finaldf
```

```python
from sklearn.linear_model import LinearRegression
clf=LinearRegression()
x=finaldf[['Mileage','Age(yrs)','BMW X5','Mercedez Benz C class']]
x
y=finaldf['Sell Price($)']
y
```

```python
clf.fit(x,y)
```

```python
clf.predict([[45000,4,0,1]])
```

```python
clf.predict([[86000,7,1,0]])
```

```python
clf.score(x,y)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dfle=df
dfle.carmodel=le.fit_transform(df.carmodel)
dfle
```

```python
x=dfle[['carmodel','Mileage','Age(yrs)']].values
x
```

```python
y=dfle['Sell Price($)']
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer([('carmodel',OneHotEncoder(),[0])],remainder='passthrough')
```

```python
clf.fit(X,y)
```
